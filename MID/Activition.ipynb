```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-29, 27, 1)
print(x)
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

## Step Function

```python
# def step(x):
#     if x > 0.5:
#         return 1
#     else:
#         return 0

# # y = []
# # for i in x:
# #     y.append(step(i))
# y = list(map(step, x))

y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

## Sigmoid Function

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

## Tanh

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

## ReLU

```python
# def relu(x):
#     if x >= 0:
#         return x
#     else:
#         return 0
# y = list(map(relu, x))

y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

## SeLU

```python
def selu(x, alpha=1.67326, scale=1.0507):
    mask = (x > 0).astype(float)
    return scale * (mask * x + alpha * (1 - mask) * (np.exp(x) - 1))
y = selu(x)
plot_graph(y,"SELU(X)")
```

## ELU

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
y = elu(x)
plot_graph(y,"ELU(X)")
```
